```python
x = torch.tensor(5.0, requires_grad=True)
f = x ** 2
f
```

requires_grad=True → PyTorch-a deyirsən ki:
“Bu dəyişənə görə törəmə (gradient) hesablanacaq”

```python
f.backward()
x.grad
```

```python
learning_rate = 0.1
with torch.no_grad():
  x -= learning_rate * x.grad  #gradient descent step
```

learning_rate = 0.1
Bu öyrənmə sürətidir
Modelin necə böyük addımlarla irəlilədiyini müəyyən edir


```python
x
```

```python
x_detached = x.detach()
x_detached -= learning_rate * x.grad
```

x_detached x in deyerlerinin kopyasidir. gradient izlemeden deyerleri goturur.

Əgər requires_grad=True olan tensoru birbaşa dəyişsən, PyTorch problem yarada bilər.
detach() manual hesablamalar ve debug üçündur

x_detached -= learning_rate * x.grad
Burda Gradient hesablanmır ve hec bir grad_fn yoxdur

detach() → “bu tensoru gradient sistemindən ayır, amma dəyərini saxla.”

```python
x_detached
```

```python
x.grad.zero_()
```

Bu kod x-in gradientini 0-a sıfırlayır. PyTorch-da gradientlər üst-üstə yığılır. Yeni backward() çağırmazdan əvvəl sıfırlamazsan, köhnə gradient-lər də əlavə olunacaq.

```python
learning_rate = 0.1
x = torch.tensor(5.0, requires_grad = True)
for iteration in range(100):
  f = x ** 2  # forward pass
  f.backward()   #backward pass
  with torch.no_grad():
    x -= learning_rate * x.grad  #gradient descent step
  print(x)
  x.grad.zero_()   #reset the gradients
```
